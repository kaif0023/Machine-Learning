```python
data["diagnosis"]=data["diagnosis"].replace({"M":1,"B":0})
```

```python
x=data.drop(columns=["diagnosis"])
y=data["diagnosis"]
```

```python
## scaling x
```

```python
sc= StandardScaler()
x_scaled=sc.fit_transform(x)
```

```python
### split the data into train and test.
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=120)
```

```python
### Now checking tha shape of train  data and  test data.
```

```python
print(y_train.shape)
print(x_train.shape)
```

```python
print(y_test.shape)
print(x_test.shape)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)  ## training
```

```python
## Evaluating  the model performance.
```

```python
y_pred=lr.predict(x_test)
```

```python
data["diagnosis"].value_counts()
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
true_positive=cm[0,0]
true_negative=cm[1,1]
false_positive=cm[1,0]
false_negative=cm[0,1]
```

```python
accuracy_score(y_test,y_pred)
```

### since the accuracy on test data is high so there is a chance of overfitting,
### so we check accuracy on training data.

```python
y_pred_train=lr.predict(x_train)
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
### from the output we can say that variance is less so this is  not a case of overfitting.
## Checking Our Next Classification i.e Recall
recall=(true_positive)/(true_positive+false_negative)
```
