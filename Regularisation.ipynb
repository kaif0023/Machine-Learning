```python
r=Ridge(alpha=0)
r.fit(x_train,y_train)
print(r.coef_)
y_ridge_pred=r.predict(x_test)
mse1=mse(y_test,y_ridge_pred)
mse1
np.sqrt(mse1)
```

##  If Alpha Is Zero Then The Ridge Model Will BEhave Like A Linear Regression Model

## Changing The Value Of Alpha 

```python
l=[0.3,0.5,1,2,5,10,50,100,150]
coef=[]
error=[]
pos=1
k=0
```

```python
plt.figure(figsize=(20,10))
for i in l:
    r=Ridge(alpha=i)
    r.fit(x_train,y_train)
    coef.append(r.coef_)
    y_ridge_pred=r.predict(x_test)
    mse1=mse(y_test,y_ridge_pred)
    rmse=np.sqrt(mse1)
    error.append(rmse)
    plt.subplot(3,3,pos)
    plt.bar(diabetes.feature_names,coef[k])
    plt.title(i)
    pos+=1
    k+=1
print(error)
```

### From The Above Barplot We Can Say That That BY INcreasing The Value Of alpha We Cut Down The Value Of Coeficient To Close To Zero

##  Using Lasso Model And Applying The Same MOdel

```python
l=[0.3,0.5,1,2,5,10,50,100,500]
cof=[]
error=[]
pos=1
k=0
plt.figure(figsize=(20,10))
for i in l:
    r=Lasso(alpha=i) 
    r.fit(x_train,y_train)
    cof.append(r.coef_.tolist())

    y_lasso_pred=r.predict(x_test)

    mse11=mse(y_test,y_lasso_pred)
    rmse=np.sqrt(mse11)
    error.append(rmse)
    plt.subplot(3,3,pos)
    plt.bar(diabetes.feature_names,cof[k])
    plt.title(i)
    pos+=1
    k+=1
print(error)
```
